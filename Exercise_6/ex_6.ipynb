```python
import matplotlib.pyplot as plt
import numpy as np

corr = df.select_dtypes(include="number").corr()

plt.figure(figsize=(14, 10))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar=True,
    square=True,
    annot_kws={"size": 10},
)

plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title("Correlation Matrix", fontsize=16)

plt.show()
```

```python
# Unstack into pairs
corr_pairs = corr.unstack()
# Drop self-correlations
corr_pairs = corr_pairs[
    corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)
]
# Remove duplicates
corr_pairs = corr_pairs[
    ~(corr_pairs.index.get_level_values(0) > corr_pairs.index.get_level_values(1))
]
# Sort by absolute correlation
sorted_pairs = corr_pairs.abs().sort_values(ascending=False)
# Show the top N pairs
top_pairs = sorted_pairs.head(20)

signed_corr = corr.unstack().loc[top_pairs.index]
num_pairs = len(top_pairs)
cols = 4
rows = int(np.ceil(num_pairs / cols))

plt.figure(figsize=(6 * cols, 5 * rows))

for i, (col_x, col_y) in enumerate(top_pairs.index):
    plt.subplot(rows, cols, i + 1)

    plt.scatter(df[col_x], df[col_y], s=8)
    plt.xlabel(col_x)
    plt.ylabel(col_y)

    c = signed_corr[(col_x, col_y)]
    plt.title(f"{col_x} vs {col_y}\nCorr = {c:.3f}")

plt.tight_layout()
plt.show()
```

### Additional Filtering
Drop column `int_rate` since it is identical to target. Drop `loan_status` due to imbalance.

```python
data = data.drop(columns=["int_rate", "loan_status"])
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

numeric_features = data.select_dtypes(include="number")

# Fill missing numeric values with median
for col in numeric_features:
    data[col] = data[col].fillna(data[col].median())

# Ordinal
ordinal_features = ["term", "emp_length"]

# Simple manual mapping for ordinals
data["term"] = data["term"].str.extract("(\d+)").astype(float)  # '36 months' -> 36
data["emp_length"] = (
    data["emp_length"]
    .replace(
        {
            "< 1 year": 0,
            "1 year": 1,
            "2 years": 2,
            "3 years": 3,
            "4 years": 4,
            "5 years": 5,
            "6 years": 6,
            "7 years": 7,
            "8 years": 8,
            "9 years": 9,
            "10+ years": 10,
        }
    )
    .astype(float)
)

# Nominal
nominal_features = [
    "home_ownership",
    "verification_status",
    "initial_list_status",
    "purpose",
    "emp_title",
    "issue_d",
]

# Fill with 'Unknown'
for col in nominal_features:
    data[col] = data[col].fillna("Unknown")
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])

# Make target int
data["target"] = data["target"].astype(int)

# Train/test split
X = data.drop(columns=["target"])
y = data["target"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
```

### Model Selection
Select xgboost for classifier. Our dataset consists of multiple features, that might have non-linear relationships among them and with the target column. Moreover, xgboost can effectively handle missing or outlier values. Moreover, xgboost can effectively handle categorical and numerical features and works good with imbalanced datasets because of boosting and regularization.

```python
model = xgb.XGBClassifier(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="logloss",
    use_label_encoder=False,
    random_state=42,
    enable_categorical=True,
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=["good", "bad"]))
```

### Evaluation
We see that our model has very good values across all metrics. The metric that I would check for such a model would be F1 because it combines both recall and precission, and is the most balanced metric. Between recall and precission, I would choose recall since I would like my model to have increased accuracy on predicting good loans even if that would mean that some good loans would be marked as bad.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

# Copy data
data_rf = data.copy()

numeric_features = data_rf.select_dtypes(include=["int64", "float64"]).columns.tolist()
numeric_features = [col for col in numeric_features if col != "target"]

scaler = StandardScaler()
data_rf[numeric_features] = scaler.fit_transform(data_rf[numeric_features])

X = data_rf.drop(columns=["target"])
y = data_rf["target"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

rf_model = RandomForestClassifier(
    n_estimators=500, max_depth=None, random_state=42, n_jobs=-1
)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=["good", "bad"]))

feature_importances = pd.Series(rf_model.feature_importances_, index=X.columns)
top15_features = feature_importances.sort_values(ascending=False).head(15)

print("Top 15 features by importance:")
print(top15_features)
```

### Mnist Dataset

```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)
print(f"x_train shape: {x_train.shape}, x_test shape: {x_test.shape}")

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

pca = PCA(n_components=300)
x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled)

ks = [5, 15, 51, 101]
for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    knn.fit(x_train_pca, y_train)
    y_pred = knn.predict(x_test_pca)
    acc = accuracy_score(y_test, y_pred)
    print(f"k={k}, Accuracy={acc:.4f}")
```

### PCA 
We cannot apply fit_transform on test_data since this set is used to simulate data that we do not have in our hands during training. Fit transform learns from the given information and applies PCA, so if we include the test set we are leaking information into the model. We cannot use test data to train the scaler, we can only use the trained scaler to transform test data and make them compatible for the knn classifier that has been trained on PCA transformed data.

```python
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2)
x_train_lda = lda.fit_transform(x_train_scaled, y_train)
x_test_lda = lda.transform(x_test_scaled)

# 2D Scatter plot
plt.figure(figsize=(16, 8))
for digit in range(10):
    plt.scatter(
        x_train_lda[y_train == digit, 0],
        x_train_lda[y_train == digit, 1],
        label=str(digit),
        alpha=0.6,
    )
plt.xlabel("LDA Component 1")
plt.ylabel("LDA Component 2")
plt.title("LDA projection of MNIST (2D)")
plt.legend()
plt.show()

fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.ravel()

x_min, x_max = x_train_lda[:, 0].min(), x_train_lda[:, 0].max()
y_min, y_max = x_train_lda[:, 1].min(), x_train_lda[:, 1].max()

for digit in range(10):
    ax = axes[digit]
    subset = x_train_lda[y_train == digit]
    ax.scatter(subset[:, 0], subset[:, 1], alpha=0.7)
    ax.set_title(f"Digit {digit}")
    ax.set_xlabel("LDA 1")
    ax.set_ylabel("LDA 2")
    # Apply shared limits
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)

plt.tight_layout()
plt.show()
```

### LDA
We observe that numbers tend to have similar shapes with others and can be easily distinguished from others. For example, `1` is easily distinguished from `0`.

```python
import umap

reducer = umap.UMAP(n_components=2, random_state=42)
x_train_umap = reducer.fit_transform(x_train_scaled)

plt.figure(figsize=(10, 8))
for digit in range(10):
    plt.scatter(
        x_train_umap[y_train == digit, 0],
        x_train_umap[y_train == digit, 1],
        label=str(digit),
        alpha=0.6,
    )
plt.xlabel("UMAP Component 1")
plt.ylabel("UMAP Component 2")
plt.title("UMAP projection of MNIST (2D)")
plt.legend()
plt.show()

x_min, x_max = x_train_umap[:, 0].min(), x_train_umap[:, 0].max()
y_min, y_max = x_train_umap[:, 1].min(), x_train_umap[:, 1].max()

fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.ravel()

for digit in range(10):
    ax = axes[digit]
    subset = x_train_umap[y_train == digit]

    ax.scatter(subset[:, 0], subset[:, 1], alpha=0.7)
    ax.set_title(f"Digit {digit}")
    ax.set_xlabel("UMAP 1")
    ax.set_ylabel("UMAP 2")

    # Shared axes
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)

plt.tight_layout()
plt.show()
```

### UMAP
From a visual inspection of the two approaches, it looks like UMAP has distinguished better the digits compared to LDA. The reason could be that LDA uses linear transformations between features while UMAP is a non-linear method. Our features that basically are pixels, might not have strong linear relationships between them.
